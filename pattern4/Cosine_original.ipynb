```python
import re
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("CPE_FAQ.xlsx")
data
```

```python
from pythainlp.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from pythainlp.corpus import thai_stopwords
```

```python
def thai_tokenizer(text):
    from pythainlp.tokenize import word_tokenize
    return word_tokenize(text, keep_whitespace=False)
```

```python
count_vect = CountVectorizer(tokenizer=thai_tokenizer, stop_words=thai_stopwords())
data_count_vect = count_vect.fit_transform(data['CPE_questions'])
```

```python
data_count_vect
```

```python
pd.DataFrame(data_count_vect.todense(), columns=count_vect.get_feature_names())
```

```python
tfidf_vect = TfidfVectorizer(tokenizer=thai_tokenizer, stop_words=thai_stopwords())
data_tfidf_vect = tfidf_vect.fit_transform(data['CPE_questions'])
```

```python
pd.DataFrame(data_tfidf_vect.todense(), columns=tfidf_vect.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_distances
```

```python
from pythainlp import word_tokenize 
from pythainlp.word_vector import * 
from sklearn.metrics.pairwise import cosine_similarity  
import numpy as np
model=get_model() 
def sentence_vectorizer(ss,dim=300,use_mean=True): 
    s = word_tokenize(ss)
    vec = np.zeros((1,dim))
    for word in s:
        if word in model.wv.index2word:
            vec+= model.wv.word_vec(word)
        else: pass
    if use_mean: vec /= len(s)
    return vec
def sentence_similarity(s1,s2):
    return cosine_similarity(sentence_vectorizer(str(s1)),sentence_vectorizer(str(s2)))
```

```python
print(sentence_similarity("วิศวคอมมีหลักสูตรอะไรบ้าง","วิศวคอมเรียนอะไรบ้าง"))
```
